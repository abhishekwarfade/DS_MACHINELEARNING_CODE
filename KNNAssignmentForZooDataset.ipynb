```python
df=df.rename(columns={"animal name":"animalname"})
df.corr()
```

```python
import statsmodels.formula.api as smf
model=smf.ols("type~animalname+predator+venomous+feathers+milk+airborne+aquatic+backbone+breathes+legs+tail+catsize",data=df).fit()
```

```python
model.resid
```

```python
plt.scatter(model.fittedvalues,model.resid)
plt.title("residuals plot",color="black",size="15")
plt.ylabel("model.resid",color="red",size=15)
plt.xlabel("model.fittedvalues",color="blue",size=15)
```

```python
import statsmodels.api as sm
```

```python
qqplot=sm.qqplot(model.resid,line="q")
plt.title("Normal QQ plot")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=4,p=2)
```

```python
knn.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("accuracy_score of training data is",ac1.round(3))
print("accuracy_score of testing data is",ac2.round(3))
```

```python
knn=KNeighborsClassifier(n_neighbors=4,p=2)

training_accuracy=[]
test_accuracy=[]
for i in range(1,101):
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=i)
    
knn.fit(X_train,Y_train)
Y_pred_train=knn.predict(X_train)
Y_pred_test=knn.predict(X_test)
from sklearn.metrics import accuracy_score
training_accuracy.append(accuracy_score(Y_train,Y_pred_train))
test_accuracy.append(accuracy_score(Y_test,Y_pred_test))
print("average Accuracy of training data is",np.mean(training_accuracy).round(3))

print("average Accuracy of test data is",np.mean(test_accuracy).round(3))
    
```

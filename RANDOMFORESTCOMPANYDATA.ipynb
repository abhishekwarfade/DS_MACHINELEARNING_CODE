```python
#label encoding
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df["Sales"]=LE.fit_transform(df["Sales"])
df["ShelveLoc"]=LE.fit_transform(df["ShelveLoc"])
df["Urban"]=LE.fit_transform(df["Urban"])
df["US"]=LE.fit_transform(df["US"])

```

```python
df
```

```python
#SQUARE ROOT TRANSFORMATIN
df.iloc[:,1:-2]=df.iloc[:,1:-2]**(1/2)
df.iloc[:,1:-2]
```

```python
#sqare transformation
df.iloc[:,1:-2]=df.iloc[:,1:-2]**(2)
df.iloc[:,1:-2]
```

```python
#min max scaler transformation
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
df.iloc[:,1:-2]=MM.fit_transform(df.iloc[:,1:-2])
df.iloc[:,1:-2]=pd.DataFrame(df.iloc[:,1:-2])
df.iloc[:,1:-2]
```

```python
df
```

```python
#data partision
```

```python
X=df.iloc[:,1:]
Y=df["Sales"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#model fitting
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=100,
                      max_samples=0.6,
                      max_features=0.7)
RFC.fit(X_train,Y_train)
```

```python
#prediction
Y_pred_train=RFC.predict(X_train)
Y_pred_test=RFC.predict(X_test)
```

```python
#model evalution
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
print("Accuracy of training data is",ac1.round(3))
ac2=accuracy_score(Y_test,Y_pred_test)
print("Accuracy of test data is",ac1.round(3))



```

```python
from sklearn.metrics import confusion_matrix
ac1=confusion_matrix(Y_train,Y_pred_train)
print("Accuracy of training data is",ac1.round(3))
ac2=confusion_matrix(Y_test,Y_pred_test)
print("Accuracy of test data is",ac1.round(3))

```

```python
#DATA TRANSFORMATION
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df["Undergrad"]=LE.fit_transform(df["Undergrad"])
df["Marital.Status"]=LE.fit_transform(df["Marital.Status"])
df["Urban"]=LE.fit_transform(df["Urban"])
```

```python
df
```

```python
#log transformation
df.iloc[:,3:]=(df.iloc[:,3:])**(1/2)
df.iloc[:,3:]
```

```python
#sqare transformation
df.iloc[:,3:]=(df.iloc[:,3:])**(2)
df.iloc[:,3:]
```

```python
#minmaxscaler
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
df.iloc[:,3:]=MM.fit_transform(df.iloc[:,3:])
pd.DataFrame(df.iloc[:,3:])
```

```python
#Split X and Y
X=df.iloc[:,0:-1]
Y=df["Taxable.Income"]
```

```python
#Data Partision
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#model fitting
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(
                      n_estimators=100,
                      max_samples=0.6,
                      max_features=0.7
                        )
RFR.fit(X_train,Y_train)
```

```python
#prediction
Y_pred_train=RFR.predict(X_train)
Y_pred_test=RFR.predict(X_test)
```

```python
#model evalution
from sklearn.metrics import mean_squared_error
mse1=mean_squared_error(Y_train,Y_pred_train)
mse2=mean_squared_error(Y_test,Y_pred_test)
print("Accuracy of training data",mse1)
print("Accuracy of test data",mse2)
print("Root mean squared error of training data is",np.sqrt(mse1).round(2))
print("Root mean squared error of test data is",np.sqrt(mse2).round(2))
print("Variance between train and test: ",mse2-mse1)
```

```python
training_error=[]
test_error=[]
for i in range(1,101):
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
    RFR.fit(X_train,Y_train)
    Y_pred_test=RFR.predict(X_test)
    Y_pred_train=RFR.predict(X_train)
    training_error.append(mean_squared_error(Y_train,Y_pred_train))
    test_error.append(mean_squared_error(Y_test,Y_pred_test))
    
```

```python
print(" Average Accuracy of training data",np.mean(training_error).round(3))
print(" Average Accuracy of test data",np.mean(test_error).round(3))
```

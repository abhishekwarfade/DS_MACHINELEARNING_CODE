```python
#model fitting 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,Y_train)
```

```python
Y_pred_train=logreg.predict(X_train)
```

```python
Y_pred_test=logreg.predict(X_test)
```

```python
Y_pred_test
```

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(Y_train,Y_pred_train)
cm2=confusion_matrix(Y_test,Y_pred_test)
print(cm1)
print(cm2)
```

```python
from sklearn.metrics import recall_score
```

```python
rs1=recall_score(Y_train,Y_pred_train)
rs2=recall_score(Y_test,Y_pred_test)
print("Sencitivity of training data is",rs1.round(3))
print("Sencitivity of testing data is",rs2.round(3))
```

```python
TN = cm1[1,1]
FP = cm1[0,1]
print("Specificity score of training data:", (TN/(TN+FP)).round(2))
TN = cm2[1,1]
FP = cm2[0,1]
print("Specificity score test data:", (TN/(TN+FP)).round(2))
```

```python
from sklearn.metrics import precision_score
ps1=precision_score(Y_train,Y_pred_train)
ps2=precision_score(Y_test,Y_pred_test)
print("precision_score of training data is",rs1.round(3))
print("precision_score of testing data is",rs2.round(3))
```

```python
from sklearn.metrics import f1_score
fs1=f1_score(Y_train,Y_pred_train)
fs2=f1_score(Y_test,Y_pred_test)
print("f1_score of training data is",rs1.round(3))
print("f1_score of testing data is",rs2.round(3))
```

```python
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("accuracy_score of training data is",rs1.round(3))
print("accuracy_score of testing data is",rs2.round(3))
```
